```python
import subprocess
import sys

def ask_airbnb(query):
    """Ask Airbnb agent."""
    code = f"""
import asyncio
from airbnb_mcp import search
asyncio.run(search("{query}"))
"""
    result = subprocess.run([sys.executable, "-c", code], 
                          capture_output=True, text=True)
    print(result.stdout)
```

```python
ask_airbnb("Find Airbnb in San Francisco for 2 adults")
```
